# Pipeline Index

## Submission Guide

This folder contains the grader-facing ML notebook submission set.

* Read `01_global_data_profiling.ipynb` first for the shared data inventory, missingness review, and exploratory context used by the pipeline notebooks.
* The canonical graded notebooks are the slug-named predictive and explanatory pairs listed below.
* Archival duplicate notebook names are marked `DO NOT SUBMIT` inside their own files and are not part of the canonical list.
